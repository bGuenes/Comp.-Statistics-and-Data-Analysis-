## 1 College achievement test, again

To find out if $90$ minutes is enough time for $90\%$ of the students to complete the test, we can use the concept of z-scores. <br>
<br>
Given: <br>
Mean $\mu$ = $70$ minutes <br>
Standard deviation $\sigma$ = $12$ minutes<br>
Desired percentage = $90\%$ <br>
<br>
First, let's find the z-score corresponding to the $90$th percentile (where $90\%$ of the data lies below it) using the standard normal distribution table or a calculator that can compute these values directly.<br>
<br>
For a standard normal distribution, the z-score corresponding to the $90$th percentile is approximately $1.28$.<br>
<br>
Next, we'll use the formula for finding the actual time value $X$ corresponding to a given z-score:<br>
<br>
$$ X = \mu + (z * \sigma)$$
Where:<br>
X = Actual time value<br>
$\mu$ = Mean<br>
$z$ = z-score<br>
$\sigma$ = Standard deviation
<br>
Substitute the values into the formula:
$$ \begin{align*}
X &= 70 + (1.28 * 12) \\
X &= 85.36 
\end{align*} $$

So, the time corresponding to the $90$th percentile is approximately $85.36$ minutes.<br>
<br>
This means that $90%$ of the students are expected to complete the test within approximately $85.36$ minutes, which is less than the $90$-minute limit. Therefore, allowing $90$ minutes should be sufficient for $90%$ of the students to complete the test.<br>
<br>

### Simulating the required time to complete the test for 1000 students with a normal distribution

```python
# Simulate the time required to complete the test for 1000 students

import numpy as np
import matplotlib.pyplot as plt

# Set the number of students
n = 10000

# Set the number of simulations
n_sim = 1000

# Set the mean and standard deviation of the time required to answer a question
mu = 70
sigma = 12

# Simulate the time required to answer each question for each student
time = np.random.normal(mu, sigma, n)

# Plot the histogram of the total time required to complete the test for each student
plt.hist(time, bins=20)
plt.xlabel('Total time required to complete the test')
plt.ylabel('Frequency')
plt.title('Histogram of the total time required to complete the test for each student')
plt.show()
```

```python
perc_of_time = np.percentile(time, [90])
print('The 90th percentile of the total time required to complete the test for each student is', round(perc_of_time[0], 2), 'minutes.')
```

The likelihood function is given by:

$$ L(x_{\text{min}}, y_{\text{min}}, x_{\text{max}}, y_{\text{max}}) = \left(\frac{1}{{x_{\text{max}} - x_{\text{min}}}}\right)^N \times \left(\frac{1}{{y_{\text{max}} - y_{\text{min}}}}\right)^N $$

The log-likelihood function is given by:

$$ \log L(x_{\text{min}}, y_{\text{min}}, x_{\text{max}}, y_{\text{max}}) = -N \left(\log(x_{\text{max}} - x_{\text{min}}) + \log(y_{\text{max}} - y_{\text{min}})\right) $$

The partial derivatives of the log-likelihood function with respect to each parameter are given by:

$$ \frac{\partial \log L}{\partial x_{\text{min}}} = \frac{N}{x_{\text{max}} - x_{\text{min}}} = 0 \\
\frac{\partial \log L}{\partial y_{\text{min}}} = \frac{N}{y_{\text{max}} - y_{\text{min}}} = 0 \\
\frac{\partial \log L}{\partial x_{\text{max}}} = -\frac{N}{x_{\text{max}} - x_{\text{min}}} = 0\\
 \frac{\partial \log L}{\partial y_{\text{max}}} = -\frac{N}{y_{\text{max}} - y_{\text{min}}} = 0 $$

Solving the system of equations, we get:

$$ x_{\text{min}} = x_{\text{max}} \\
y_{\text{min}} = y_{\text{max}} $$

This means that the maximum likelihood estimates of the parameters are:

$$ x_{\text{min}} = \min(x_n)\\
y_{\text{min}} = \min(y_n)\\
x_{\text{max}} = \max(x_n)\\
 y_{\text{max}} = \max(y_n) $$

In other words, the maximum likelihood estimates of the parameters are the minimum and maximum observed values of $ x $ and $ y $.
