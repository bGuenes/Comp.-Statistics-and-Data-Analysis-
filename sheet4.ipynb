## 4 Random Walks

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.linspace(10, 1000, 100)
runs = 100
```

```python
fig, ax = plt.subplots(1, 1)

for i in n:
    x_endpoint = []
    y_endpoint = []
    x_var_list = []
    y_var_list = []

    for j in range(runs):
        x = np.random.normal(0, 1, int(i))
        y = np.random.standard_cauchy(int(i))

        x = np.sum(x)
        y = np.sum(y)

        x_endpoint.append(x)
        y_endpoint.append(y)

    x_var_list = np.var(x_endpoint)
    y_var_list = np.var(y_endpoint)

    ax.scatter(i, x_var_list, color='blue', s=5)
    ax.scatter(i, y_var_list, color='red', s=5)


ax.set_xscale("log")
ax.set_yscale("log")
plt.xlabel('Number of steps')
plt.ylabel('Variance')
plt.legend(['Brownian Motion', 'Lévy Motion'])
plt.show()

```
